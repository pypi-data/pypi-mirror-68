```python
%pylab inline
from pycheops.funcs import xyz_planet
from numpy import linspace
t = linspace(0,1,1000)
sini = 0.998
ecc = 0.34
omdeg = 45

x, y, z = xyz_planet(t,0,1,sini,ecc,omdeg)

fig, ax = subplots(2,2,figsize=(8,7))

ax[0][0].plot(x, y)
ax[0][0].scatter(x[0], y[0])
ax[0][0].set_xlabel('x')
ax[0][0].set_ylabel('y')
ax[0][0].set_xlim(-1.5,1.5)
ax[0][0].set_ylim(-1.5,1.5)


ax[0][1].plot(x, z)
ax[0][1].scatter(x[0], z[0])
ax[0][1].set_xlabel('x')
ax[0][1].set_ylabel('z')
ax[0][1].set_xlim(-1.5,1.5)
ax[0][1].set_ylim(-1.5,1.5)

ax[1][0].plot(z, y)
ax[1][0].scatter(z[0], y[0])
ax[1][0].set_xlabel('z')
ax[1][0].set_ylabel('y')
ax[1][0].set_xlim(-1.5,1.5)
ax[1][0].set_ylim(-1.5,1.5)

r = sqrt(x**2+y**2+z**2)
ax[1][1].plot(t, r)
ax[1][1].set_xlabel('t')
ax[1][1].set_ylabel('r')
ax[1][1].set_xlim(0,1)
ax[1][1].set_ylim(0,1.5)

fig.tight_layout()
```

### Lambertian phase function

```python
fig, ax = subplots(1,2,figsize=(8,3.5))

bet = arccos(z/r)
ax[0].plot(t, bet)
ax[0].set_xlabel('t')
ax[0].set_ylabel(r'$\beta$')
ax[0].set_xlim(0,1)
ax[0].set_ylim(0,pi)

Phi_L = (sin(bet) + (pi-bet)*cos(bet))/pi
ax[1].plot(t, Phi_L)
ax[1].set_xlabel('t')
ax[1].set_ylabel(r'$\Phi_L$')
ax[1].set_xlim(0,1)
ax[1].set_ylim(0,1)

fig.tight_layout()
```
