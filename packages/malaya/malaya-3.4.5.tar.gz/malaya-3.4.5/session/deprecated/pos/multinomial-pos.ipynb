```python
import json
import re
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
def process_string(string):
    string = re.sub('[^A-Za-z0-9\-\/ ]+', ' ', string).split()
    return [y.strip() for y in string]
```

```python
with open('pos-data-v3.json','r') as fopen:
    dataset = json.load(fopen)
```

```python
texts, labels = [], []
for i in dataset:
    try:
        texts.append(process_string(i[0])[0].lower())
        labels.append(i[-1])
    except Exception as e:
        print(e, i)
```

```python
target = LabelEncoder().fit_transform(labels)
bow_chars = CountVectorizer(ngram_range=(2, 4), analyzer='char').fit(texts)
vectors = bow_chars.transform(texts)
vectors.shape
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(vectors, target, test_size = 0.1)
del vectors
```

```python
from sklearn import metrics
```

```python
train_Y[:10]
```

```python
multinomial = MultinomialNB().fit(train_X, train_Y)
print(metrics.classification_report(train_Y, multinomial.predict(train_X), target_names = np.unique(labels)))
```

```python
import pickle
with open('multinomial-pos.pkl','wb') as fopen:
    pickle.dump(multinomial,fopen)
with open('bow-pos.pkl','wb') as fopen:
    pickle.dump(bow_chars,fopen)
```
