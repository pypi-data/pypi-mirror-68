# Create a cluster

```python
from aws_jupyter.create_cluster import create_cluster

create_cluster({
    "count": 2,
    "name": "new-cluster",
    "type": "r3.large",
    "spot": 1.0,
})
```

# Check if the cluster is ready

```python
from aws_jupyter.check_cluster import check_status_and_init
from time import sleep

while not check_status_and_init({"name": "new-cluster",}):
    sleep(2)
```

# Terminate a cluster

```python
from aws_jupyter.terminate_cluster import terminate_cluster

terminate_cluster({
    "name": "new-cluster",
})
```
