```python
llabels = ['noEnergy', 'music', 'noise', 'female', 'male']

lrec = []
for f in sorted(glob.glob('./surrey3/*.csv')):
    #print(f)
    base, _ = os.path.splitext(os.path.basename(f))
    df = pd.read_csv(f, sep='\t')
    #print(df)
    dur = str(timedelta(0, df.stop.max()))
    rec = [base, dur]
    df['duration'] = df.stop - df.start
    for label in llabels:
        rec.append(df[df.labels == label].duration.sum())
    lrec .append(tuple(rec))
```

```python
df = pd.DataFrame.from_records(lrec, columns=['filename', 'total_duration(h:m:s)'] + llabels)
df['female_percentage'] = df.female * 100. / (df.female + df.male)
df
```

```python
df.to_excel('./surrey_inaspeechsegmenter0-3-3.xlsx', index=False)
```
